```python
from arch import arch_model
import bacen
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests as gc
```

```python
di  = bacen.get(12,(2018,8,31),(2021,3,10))
di.index = pd.to_datetime(di.index)
di.head()
```

```python
news = pd.read_csv('scrapy/news.csv')
news = news.dropna()
news = news.iloc[::-1]
news.head()
```

```python
day_start    = 1
day_end      = 3
month_start  = 4
month_end    = 6
year_start   = 7
year_end     = 11
hour_start   = 12
hour_end     = 14
minute_start = 15
minute_end   = 17
```

```python
lis_t = []
for date in news['date']:
    day    = int(date[day_start:day_end])
    month  = int(date[month_start:month_end])
    year   = int(date[year_start:year_end])
    hour   = int(date[hour_start:hour_end])
    minute = int(date[minute_start:minute_end])
    date_time = dt.datetime(year,month,day,hour,minute)
    time_stamp = pd.Timestamp(date_time)
    lis_t.append(time_stamp)
```

```python
news.index = lis_t
news = news['link']
news.head()
```

```python
lis_t1 = []
for date in di.index:
    lis_t1.append(date + dt.timedelta(hours=18))
```

```python
di.index = lis_t1
di.head()
```

```python
volume = []
for i in range(len(di)):
    today =0
    if i == 0:
        for ind in news.index:
            if ind < di.index[i]:
                today += 1
    else:
        for ind in news.index:
            if ind > di.index[i-1] and ind < di.index[i]:
                today += 1
    volume.append(today)
```

```python
di['volume'] = volume
```

```python
di
```

```python
di['log_ret'] = np.log(di['valor']) - np.log(di['valor'].shift(1))
```

```python
fig, ax = plt.subplots()
ax.plot(di['log_ret'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(di['volume'])
plt.show()
```

```python
am = arch_model(di['log_ret'][1:])
res = am.fit()
print(res.summary())
```

```python
di['cvol'] = res.conditional_volatility
```
